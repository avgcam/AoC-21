```python
def readlines():
	x = []
	with open('input2.txt') as file:
		for line in file:
			dir, n = line.rstrip().split(" ")
			n = int(n)
			x.append((dir, n))
		x = tuple(x)
	return x

def part1(x):
	depth = 0
	hor_pos = 0

	for (dir, n) in x:
		if dir == 'forward':
			hor_pos += n
		elif dir == 'up':
			depth -= n
		elif dir == 'down':
			depth += n
	return hor_pos*depth

def main():
    print(part1(readlines()))
    #part2()


if __name__ == '__main__':
    main()
```

```python
horizontal, depth, aim = 0, 0, 0
with open('input2.txt') as file:
  for line in file:
    dir, magnitude = line.rstrip().split(' ')
    magnitude = int(magnitude)
    if dir == 'forward':
      horizontal += magnitude
      depth += aim * magnitude
    elif dir == 'down':
      aim += magnitude
    else:
      aim -= magnitude
print(horizontal * depth)
```
